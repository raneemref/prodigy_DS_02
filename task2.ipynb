```python
import pandas as pd

# Load the dataset
data = pd.read_csv("C:/Users/HP/Downloads/test.csv")

# Display the first few rows of the dataset
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
missing_values
```

```python
# Fill missing values in Age with the median age
data['Age'].fillna(data['Age'].median(), inplace=True)

# Fill the missing value in Fare with the median fare
data['Fare'].fillna(data['Fare'].median(), inplace=True)

# Drop the Cabin column
data.drop(columns='Cabin', inplace=True)

# Verify that there are no more missing values
missing_values_after = data.isnull().sum()
missing_values_after
```

```python
# Summary statistics for numerical columns
print(data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Distribution of Fare
plt.figure(figsize=(10, 6))
sns.histplot(data['Fare'], kde=True)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count plot for Pclass
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', data=data)
plt.title('Passenger Class Counts')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()

# Count plot for Sex
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', data=data)
plt.title('Gender Counts')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

# Count plot for Embarked
plt.figure(figsize=(10, 6))
sns.countplot(x='Embarked', data=data)
plt.title('Port of Embarkation Counts')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show()
```

```python
# Pairplot to explore relationships between numerical variables
sns.pairplot(data[['Age', 'Fare', 'Pclass', 'SibSp', 'Parch']])
plt.show()

# Box plot for Fare by Pclass
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=data)
plt.title('Fare by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()

# Box plot for Age by Sex
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sex', y='Age', data=data)
plt.title('Age by Gender')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```
